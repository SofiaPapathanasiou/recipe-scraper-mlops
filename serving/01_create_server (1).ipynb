### Set up the server

````python
from chi import server, context, lease
import os

context.version = "1.0"
context.choose_project()
context.choose_site(default="CHI@UC")
````

````python
# runs in Chameleon Jupyter environment
l = lease.get_lease(f"proj22_serve_gpuv100")
l.show()
````

````python
# runs in Chameleon Jupyter environment
username = os.getenv('USER') # all exp resources will have this prefix
s = server.Server(
    f"node-serve-model-{username}",
    reservation_id=l.node_reservations[0]["id"],
    image_name="CC-Ubuntu24.04-CUDA"
)
s.submit(idempotent=True)
````

````python
s.associate_floating_ip()
````

````python
s.refresh()
s.check_connectivity()
````

````python
s.refresh()
s.show(type="widget")
````

### Create Docker Image and launch jupyter on it to run experiments

``` docker volume create recipe-nlg```

``` cd /recipe-scraper-mlops/serving/docker```

```docker compose -f docker-compose-data.yaml up -d```

Then check to see if the data has been loaded:

```docker run --rm -it -v recipe-nlg:/data alpine ls -l /data/eval.jsonl```

To see the first 5 entries in the dataset: 

```docker run --rm -it -v recipe-nlg:/data python:3.11 bash -lc 'head -n 5 /data/eval.jsonl'```

Build the jupyter container image: 

```docker build -t jupyter-onnx-base -f Dockerfile.jupyter-onnx-base .```

Then launch the notebooks: 

``` docker run  -d --rm  -p 8888:8888 \
